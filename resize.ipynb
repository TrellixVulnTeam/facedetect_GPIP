```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read image
img = cv2.imread('./data/parrot.jpg')
# convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
img_re = cv2.resize(gray, (600,300), cv2.INTER_AREA)
img_re.shape
```

```python
plt.imshow(img_re, cmap='gray')
plt.show()
```

```python
#shrink
img_en = cv2.resize(gray, (1200,600), cv2.INTER_CUBIC)
img_en.shape
```

```python
plt.imshow(img_en, cmap='gray')
plt.show()
```
